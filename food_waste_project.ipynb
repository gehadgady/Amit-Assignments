```python
# checking for outliers
num_cols=df.select_dtypes(include=["float","int"])
num_cols
```

```python
def boxPlots (num_cols):
  plt.figure(figsize=(10,5))
  for i,col in enumerate (num_cols.columns, start=1):
    plt.subplot(8,1,i)
    # sns.boxplot(x=col,data=num_cols[col])
    sns.boxplot(x=num_cols[col],color="red")
    plt.title(col)
  plt.tight_layout(pad=3.0)  # Add padding between plots
  plt.show()
```

```python
boxPlots(num_cols)
```

```python
# outliers are in meals_served , temperature_C , food_waste_kg
```

```python
# what is hte zscore
# why in that function in the remove-> it looks like it removes all data
# on what basis we choose the threshold
def IQR_handle_outliers_fn(allData, columns,threshold):
  total_outliers = 0
  for col in columns:
    Q1 = allData[col].quantile(0.25)
    Q3 = allData[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR

    # calculating the outliers
    outliers = allData[(allData[col] < lower_bound) | (allData[col] > upper_bound)]

    outlier_count = len(outliers)
    total_outliers += outlier_count
    print(f"for the column {col} there are {outlier_count} outliers")

    #now we replace the outliers in each column in the my data
    allData[col]= np.clip(allData[col], lower_bound, upper_bound)

  return allData
```

```python
print(len(df["meals_served"]))
print(len(df["temperature_C"]))
print(len(df["staff_experience"]))
```

```python
ouliers_cols=["meals_served","temperature_C","food_waste_kg"]
```

```python
df = IQR_handle_outliers_fn(df,ouliers_cols,1.5)
```

```python
# updated,
# reducing the threshold to get the more extreme outliers to my data
df = IQR_handle_outliers_fn(df,ouliers_cols,.75)
```

```python
def countplots(data):
  data=data.select_dtypes("category")
  print(data)

  num_features = len(data.columns)  # Count numeric columns
  rows = (num_features + 1) // 2  # Create enough rows (2 plots per row)
  cols = 2  # Fixed number of columns

  plt.figure(figsize=(10,5))
  for index,col in enumerate(data.columns,1):
    # plt.subplot(1,1,index)
    plt.subplot(rows,cols,index)
    plt.title(f"Count plot plot {col}")
    sns.countplot(x=col,data=data,palette="mako",hue=col)
  plt.tight_layout()
  plt.show()
```

# Pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, RobustScaler
from sklearn.ensemble import GradientBoostingRegressor
```

```python
dtypes= df.dtypes
dtypes
```

```python
num_cols = ["past_waste_kg","day_of_week","humidity_percent","temperature_C","kitchen_staff","meals_served"]
cat_cols = ['staff_experience','waste_category',"special_event"]
```

```python
preprocessor = ColumnTransformer([
    ("num_cols", Pipeline([
        ("imputer", SimpleImputer(strategy="mean")),
        ("scaler", RobustScaler())
    ]), num_cols),
   ("cat_cols", Pipeline([
       ("imputer", SimpleImputer(strategy="most_frequent")),
       ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)),
   ]), cat_cols)
])
```

```python
lr_pipeline_model = Pipeline([
    ("preprocessor", preprocessor),
    ("model", LinearRegression())
])
```

```python
x=df.drop(columns="food_waste_kg")
y=df["food_waste_kg"]
from sklearn.model_selection import train_test_split
x_train22,x_test22,y_train22,y_test22=train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
lr_pipeline_model.fit(x_train22, y_train22)
```

```python
y_pred_ppline=lr_pipeline_model.predict(x_test22)
```

```python
r2_score(y_test22,y_pred_ppline)
```

```python
mean_absolute_error(y_test22,y_pred_ppline)
```
